```python
import pandas as pd
import sqlite3

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('./data/combined_df.csv')

# Assuming you want to insert into a table called 'energy_data'
table_name = "energy_data"

# Generate SQL INSERT statements from the DataFrame
sql_inserts = "\n".join([
    f"INSERT INTO {table_name} ({', '.join(df.columns)}) VALUES ({', '.join(['?' for _ in df.columns])});"
    for _ in df.itertuples(index=False)
])

# Save the INSERT statements to a .sql file
with open('./data/inserts.sql', 'w') as file:
    file.write(sql_inserts)
```

```python
import pandas as pd
import sqlite3

# Load your CSV file into a pandas DataFrame
df = pd.read_csv('./data/combined_df.csv')

# Create a new SQLite database (or connect to one if it already exists)
conn = sqlite3.connect('eirgrid_data.db')

# Convert the DataFrame to SQL, creating a new table 'energy_data' in the SQLite database
df.to_sql('energy_data', conn, if_exists='replace', index=False)

# Commit changes and close the connection
conn.commit()
conn.close()
```
